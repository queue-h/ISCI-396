```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
```

```python
x = np.linspace(0, 2 * np.pi, 6)
y = np.sin(x)
x_plot = np.linspace(0, 2 * np.pi, 1000)
```

```python
plt.scatter(x, y, label = 'data')
plt.plot(x_plot, np.sin(x_plot), label = 'truth')
plt.plot(x_plot, interp1d(x, y, kind='cubic')(x_plot), label = 'cubic')
plt.plot(x_plot, GaussianProcessRegressor().fit(x[:, None], y).predict(x_plot[:, None]), label = 'guassian')
plt.legend()
plt.show()
```
